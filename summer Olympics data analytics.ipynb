## 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python

goldMedal = df.loc[df['Medal'] == 'Gold']
a=goldMedal.Sport.value_counts().reset_index(name='Medal')

a.plot(x="index", y="Medal",kind="bar",figsize = (20,5))
```

## 6. In which year India won first Gold Medal in Summer Olympics?

```python
goldMedal = df.loc[(df['Medal'] == 'Gold') & (df['Country'] == 'IND')]
a=goldMedal.Year.value_counts().reset_index(name='Medal')
print("First gold medal India won in year :",a.loc[0][0])
```
